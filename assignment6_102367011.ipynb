```python
import numpy as np
import matplotlib.pyplot as plt


M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 400)

y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r'$y = Mx^2$', color='b', linestyle='-')
plt.plot(x, y2, label=r'$y = M \sin(x)$', color='r', linestyle='--')

plt.title("Plot of y = Mx^2 and y = M sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Computer'],
    'Score': [85, 90, 78, 88, 95]
}

df = pd.DataFrame(data)

# Set the style
sns.set_style("whitegrid")

plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Subject', y='Score', data=df, hue='Subject', palette='husl', legend=False)

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title("Scores by Subject", fontsize=14, fontweight='bold')
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Scores", fontsize=12)

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set roll number as seed (replace with actual roll number)
roll_number = 102367011  # Change this to your roll number
np.random.seed(roll_number)

# Generate dataset of 50 values
data = np.random.randn(50)

# Create subplots (2x2 layout)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line plot showing cumulative sum
axes[0, 0].plot(np.cumsum(data), marker='o', linestyle='-')
axes[0, 0].set_title("Cumulative Sum")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)

# Scatter plot with random noise
axes[0, 1].scatter(range(50), data, c='red', marker='x')
axes[0, 1].set_title("Scatter Plot with Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Random Values")
axes[0, 1].grid(True)

# Hide the remaining empty subplots
axes[1, 0].axis("off")
axes[1, 1].axis("off")

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")

sns.set_style("darkgrid")

plt.figure(figsize=(10, 5))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', color='b')
plt.title("Total Profit Over Months", fontsize=14, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Profit", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

plt.figure(figsize=(12, 6))
for column in df.columns[1:-1]: 
    sns.lineplot(x=df['month_number'], y=df[column], marker='o', label=column)

plt.title("Sales of Different Products Over Months", fontsize=14, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Units Sold", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Products")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

plt.figure(figsize=(12, 6))
df.set_index("month_number").plot(kind="bar", figsize=(12, 6), colormap="tab10", width=0.8, edgecolor="black")
plt.title("Comparison of All Features Across Months", fontsize=14, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Features", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
