```python
#Q.1
import numpy as np
gfg=np.matrix('[4,1,9;12,3,1;4,5,6]')
sum_all_elements=np.sum(gfg)
print("sum of all elements:",sum_all_elements)

row_sum=np.sum(gfg, axis=1)
print("sum of all elements row wise:",row_sum)

column_sum=np.sum(gfg, axis=0)
print("sum of all elements column wise:",column_sum)
```

```python
#Q.2(a)
array_a=np.array([10,52,62,16,16,54,453])
sorted_array=np.sort(array_a)
print("sorted array:",sorted_array)

sorted_indices=np.argsort(array_a)
print("Indices of sorted array:",sorted_indices)

four_smallest_elements=np.sort(array_a)[:4]
print("The four smallest elements are:",four_smallest_elements)

five_largest_elements=np.sort(array_a)[-5:]
print("The five largest elements are:",five_largest_elements)


```

```python
#Q.2(b)
array_b=np.array([1.0,1.2,2.2,2.0,3.0,2.0])

integer_elements=array_b[array_b==array_b.astype(int)]
print("Integer elements only:",integer_elements)

float_elements=array_b[array_b!=array_b.astype(int)]
print("Float elements only:",float_elements)

```

```python
#Q.3
import numpy as np


initials = ['S', 'S']
ascii_sum = sum(ord(ch) for ch in initials)
X = ascii_sum


sales = np.array([X, X + 50, X + 100, X + 150, X + 200])
print("Sales dataset:", sales)


tax_rate = ((X % 5) + 5) / 100
print("Personalized tax rate:", tax_rate)


sales_after_tax = sales * (1 + tax_rate)
print("Sales after tax:", sales_after_tax)


discounted_sales = np.where(sales < X + 100, sales * 0.95, sales * 0.90)
print("Discounted sales:", discounted_sales)


weekly_sales = np.tile(sales, (3, 1))
print("Weekly sales data (initial):\n", weekly_sales)


weeks = np.arange(3).reshape(-1, 1)  
weekly_sales = weekly_sales * (1 + 0.02 * weeks)
print("Weekly sales data (after 2% increase per week):\n", weekly_sales)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-10, 10, 100)

# Compute y values for each function
y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

# Plot the functions
plt.figure(figsize=(12, 10))

# Y = x^2
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='y = x^2',color='r')
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Y = sin(x)
plt.subplot(2, 2, 2)
plt.plot(x, y2, label='y = sin(x)',color='purple')
plt.title('y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Y = exp(x)
plt.subplot(2, 2, 3)
plt.plot(x, y3, label='y = exp(x)',color='b')
plt.title('y = exp(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Y = log(|x| + 1)
plt.subplot(2, 2, 4)
plt.plot(x, y4, label='y = log(|x| + 1)',color='g')
plt.title('y = log(|x| + 1)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
